# Exploring/visualizing energy and temperature data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
pd.options.display.max_rows = 999
```

## Loading files

```python
train = pd.read_csv('train.csv',parse_dates = ["DATE"])
```

```python
train.set_index('DATE',inplace=True)
```

```python
train.head()
```

## Plot the training energy data over time

```python
#matplot
plt.rcParams["figure.figsize"] = (8,5)
x = train.index
y = train.mw
plt.plot(x,y)
plt.xlabel('Time')
plt.ylabel('Energy')
plt.set_figsize=(25,15)
plt.show()
```

```python
#pandas
train.mw.plot(figsize=(8,5),color='black').set_ylabel('MW Hours')
```

## Plotting one week of the training energy data

```python
week_data = train['2017-02-06':'2017-02-11']
```

```python
week_data.mw.plot(figsize=(12,5)).set_ylabel('MW Hours')
```

```python
# Increment of energy activity on weekend
```

```python
#### Plotting an specific day
```

```python
daily_data = train['2017-02-09 00:00:00':'2017-02-09 23:59:59']
```

```python
daily_data.mw.plot(figsize=(12,5)).set_ylabel('MW Hours')
```

```python
#Increment of energy activity as hour of the day increases
```

## Plot the training temperature data over time

```python
train.HourlyDryBulbTemperature.plot(figsize=(8,5)).set_ylabel('Temperature (Degrees F)')
```

```python
# Overlaping 2 variables
train.HourlyDryBulbTemperature.plot(figsize=(8,5))
train.mw.plot(figsize=(8,5)).set_ylabel('Temperature (Degrees F)')
```
